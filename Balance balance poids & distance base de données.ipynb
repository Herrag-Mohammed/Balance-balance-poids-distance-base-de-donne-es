```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from matplotlib import pyplot as plt
```

```python
balance_data = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-\
databases/balance-scale/balance-scale.data',sep= ',', header = None)
```

```python
print("dataset : ")
print(balance_data.head())
```

```python
print("Nombre d'instances :",balance_data.shape[0])
print("Nombre d'attributs :",balance_data.shape[1])
```


Tout d'abord, nous devons séparer la variable cible des attributs de l'ensemble de données.


```python
X = balance_data.values[:, 1:5]
Y = balance_data.values[:,0]
```


L'étape suivante consiste à diviser l'ensemble de données à des données d'entraînement et de test.


```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, random_state = 100)
```

```python
clf_gini=DecisionTreeClassifier(random_state = 100,criterion="gini",
            max_depth = 3, min_samples_leaf = 5)
clf_gini.fit(X_train,y_train)
```

```python
plt.figure(figsize=(18,8))
plot_tree(clf_gini,filled=True)
plt.show
```

```python
clf_entropy=DecisionTreeClassifier(random_state = 100,criterion="entropy",
            max_depth = 3, min_samples_leaf = 5)
clf_entropy.fit(X_train,y_train)
```
